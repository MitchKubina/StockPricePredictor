```python
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period="max")
sp500
```

```python
sp500.plot.line(y = "Close", use_index = True)
```

```python
sp500.columns
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500.columns
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
```

```python
sp500["Growth"] = (sp500["Close"] < sp500["Tomorrow"]).astype(int)
sp500
```

```python
sp500 = sp500.loc["1990-01-01":].copy()
sp500
```

```python
from sklearn.model_selection import train_test_split

import xgboost as xgb
```

```python
x = sp500[["Open","High","Low","Close","Volume"]]
y = sp500["Growth"]
```

```python
train_x, test_x, train_y, test_y = train_test_split(x,y, random_state = 20, test_size = 0.2, shuffle = True)
```

```python
train_x
```

```python
test_x
```

```python
train_y
```

```python
test_y
```

```python
from xgboost import XGBRegressor
```

```python
model = XGBRegressor()

model.fit(train_x, train_y, verbose = False)
```

```python
y_predictions = model.predict(test_x)
y_predictions
```

```python
test_y
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
#Not great
mean_absolute_error(test_y, y_predictions)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest_model = RandomForestRegressor()
```

```python
forest_model.fit(train_x,train_y)
```

```python
forest_predictions = forest_model.predict(test_x)
```

```python
forest_predictions
```

```python
mean_absolute_error(test_y, forest_predictions)
```
